-->Credit Default Prediction

The data set consists of 2000 samples from each of two categories. Five variables are1. Income 2. Age  3. Loan 4. Loan to Income (engineered feature)5. Default

Dataset: https://github.com/ybifoundation/Dataset/raw/main/Credit%20Default.csv

```python
import pandas as pd
```

```python
credit=pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Credit%20Default.csv')
```

```python
credit.columns
```

```python
y=credit['Default']
```

```python
X=credit[['Income', 'Age', 'Loan', 'Loan to Income']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=10)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
